```python
import pandas as pd
import os
from scipy.signal import find_peaks
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('export.csv')
```

```python
arr = df.close.to_numpy()
```

```python
arr[:1000]
peaks = find_peaks(arr[:10000])
```

```python
dot_y = []
for i, point in enumerate(arr[:10000]):
    if i in peaks[0]:
       dot_y.append(point)
```

```python
len(peaks[0]), len(dot_y)
```

```python
plt.plot(arr[:10000])
plt.plot(peaks[0], dot_y, 'r.')
plt.show()
```
